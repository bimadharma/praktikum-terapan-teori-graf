##Deskripsi
Pemerintah kota menggunakan algoritma Dijkstra untuk mengoptimalkan jaringan transportasi di wilayah mereka. Mereka mencari jalur terpendek antara titik awal (misalnya gudang) dan titik tujuan (misalnya pusat distribusi) dalam jaringan jalan kota. Dengan menggunakan jalur terpendek yang ditemukan, mereka dapat mengatur arus lalu lintas dan meminimalkan waktu perjalanan serta biaya transportasi bagi kendaraan yang beroperasi di wilayah tersebut.

Case Study: Bagimana menemukan jalur terpendek untuk menuju suatu tempat?

Case Study: Tentukan lintasan terpendek dari A ke Z

```python
import heapq

def dijkstra(graph, start, end):
    # Simpan jarak terpendek dari start ke setiap node
    distances = {node: float('inf') for node in graph}
    distances[start] = 0

    # Simpan jalur terpendek dari start ke setiap node
    shortest_paths = {node: [] for node in graph}
    shortest_paths[start] = [start]

    # Priority queue untuk menyimpan node yang akan dieksplorasi selanjutnya
    priority_queue = [(0, start)]

    while priority_queue:
        current_distance, current_node = heapq.heappop(priority_queue)

        # Jika ditemukan jarak yang lebih pendek, perbarui jarak dan jalur
        if current_distance > distances[current_node]:
            continue

        for neighbor, weight in graph[current_node].items():
            distance = current_distance + weight

            # Jika jalur baru lebih pendek, perbarui jarak dan jalur
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                shortest_paths[neighbor] = shortest_paths[current_node] + [neighbor]
                heapq.heappush(priority_queue, (distance, neighbor))

    return distances[end], shortest_paths[end]

# Grafik berbobot baru
graph = {
    'A': {'B': 8, 'C': 2},
    'B': {'C': 5, 'D': 10},
    'C': {'B': 5, 'D': 7, 'E': 9},
    'D': {'E': 4, 'Z': 4},
    'E': {'D': 4, 'Z': 5},
    'Z': {}
}

start_node = 'A'
end_node = 'Z'

shortest_distance, shortest_path = dijkstra(graph, start_node, end_node)
print(f"Lintasan terpendek dari {start_node} ke {end_node}:")
print("Jarak:", shortest_distance, "meter")
print("Rute:", ' -> '.join(shortest_path))
```

##Penyelesaian:
1 Tentukan titik awal dan titik tujuan sebelum proses:
Titik awal = A
Titik tujuan = Z

2 Akumulasikan jarak minimal dan simpan ke titik berikutnya. Lakukan dari titik awal sampai titik tujuan:
- Titik A ke titik C = 2 meter
- Titik C ke titik D = 7 meter
- Titik D ke titik Z = 4 meter

Jadi, total jarak minimal dari titik awal (A) ke titik tujuan (Z) adalah 2 + 7 + 4 = 13 meter.
Dengan jalur yang sesuai,  yaitu A-C-D-Z.


##Pertanyaan konteks:

Directed/Undirected?

Jawab: Directed( A -> C -> D -> Z). Graf yang diberikan adalah graf berarah karena setiap tepi memiliki arah yang ditentukan.

Weighted/Unweighted

Jawab: Weighted(13 Meter). Graf tersebut adalah graf berbobot karena setiap tepi memiliki bobot yang menentukan jarak antar simpul.

Sparse/Dense?

Jawab: Sparse. Graf tersebut tidak terlalu padat karena jumlah tepinya relatif kecil dibandingkan dengan jumlah simpulnya.

Karakteristik masalah tipe di atas adalah Shortest-Path-Problem, terutama menggunakan algoritma Dijkstra untuk menemukan lintasan terpendek dari satu simpul ke simpul lain dalam graf berarah dan berbobot.
